```python
def categorize_athelete(row):
    if row['height_cm'] < 175 and row['weight_kg'] <70:
        return 'Light_Weight'
    elif row['height_cm'] <185 or row['weight_kg'] <= 80:
        return 'Middle_Weight'
    else:
        return 'Heavy_Weight'
    
bios['Category'] = bios.apply(categorize_athelete, axis=1)
```
